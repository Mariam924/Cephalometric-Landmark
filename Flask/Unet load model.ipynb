```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import cv2
from tensorflow.keras.models import load_model
import time
```

```python
unet_dice_model = load_model('Midterm Model/Midterm_Model.h5', compile=False)
unet_focalT_model = load_model('Midterm_Model_FT.h5', compile=False)
```

```python
masks_dict = np.load('true_masks.npz')
```

```python
masks_dict['400.jpg']
```

```python
# preprocess image to be ready for prediction
def unet_preprocess(image):
    # step 1: take one channel(grayscale) and normalize
    image_norm = image[:,:,0] / 255.0

    # step 2: resize
    image_resized = cv2.resize(image_norm, (256,256))   # dimensions (256,256)
    image_ready = np.reshape(image_resized, (1, image_resized.shape[0], image_resized.shape[1], 1)) # (1,256,256,1) --> (batch, dimensions)

    return image_ready
```

```python
# return image with landamarks drawn on it
def draw_landmarks(image, pred_mask, true_mask, all_in_one=True):
    image_convert = image[:,:,0].astype(np.float32) # cv2 accepts dimension(256,256) and float32
    # convert to RGB
    image_rgb = cv2.cvtColor(image_convert, cv2.COLOR_GRAY2RGB) 

    if all_in_one:                                        # draw 2 masks on same image
        image_rgb[:,:,0] = image_rgb[:,:,0] + true_mask
        image_rgb[:,:,1] = image_rgb[:,:,1] + pred_mask[:,:,0]
        return image_rgb
    
    im_copy = image_rgb.copy()
    image_rgb[:,:,1] = image_rgb[:,:,1] + true_mask     # true mask
    im_copy[:,:,0] = im_copy[:,:,0] + pred_mask[:,:,0]   # predicted mask
    return image_rgb, im_copy
```

```python
# unet dice loss version
def unet_dice(image, image_name):     # received image is of dimensions (1935,2400,3)
    # step 1: preproccess it
    image_preprocessed = unet_preprocess(image)   # (1,256,256,1)

    # step 2: prediction
    pred_mask = unet_dice_model.predict(image_preprocessed)  # pred mask dimensions (1,256,256,1)

    true_mask = masks_dict[image_name]
    image_w_landmarks = draw_landmarks(image_preprocessed[0], pred_mask[0], true_mask, True)
    image_w_landmarks_resize = cv2.resize(image_w_landmarks, (1935,2400))
    return image_w_landmarks_resize
```

```python
# unet focal tversky version
def unet_focalt(image, image_name):     # received image is of dimensions (1935,2400,3)
    # step 1: preproccess it
    image_preprocessed = unet_preprocess(image) # (1,256,256,1)

    # step 2: prediction
    pred_mask = unet_focalT_model.predict(image_preprocessed)  # pred mask dimensions (1,256,256,1)

    true_mask = masks_dict[image_name]
    image_w_landmarks = draw_landmarks(image_preprocessed[0], pred_mask[0], true_mask, True) # final image with landmarks (256,256,3)
    image_w_landmarks_resize = cv2.resize(image_w_landmarks, (1935,2400))
    return image_w_landmarks_resize
```

```python
image = cv2.imread('Kaggle dataset/cepha400/400.jpg')
```

```python
s = time.time()
labeled_image = unet_dice(image, '400.jpg')
e = time.time()
print(e-s)
```
